```python
# 2. Import vLLM
from vllm import LLM, SamplingParams
import time

# 3. Define prompts
prompts = [
    "The sun seems warmer today",
    "I hope we can go out ",
    "Are you listening?",
]

# 4. Define sampling parameters
sampling_params = SamplingParams(
    temperature=0.8,
    top_p=0.95,
    max_tokens=50,  # number of tokens to generate
)

# 5. Run generation
llm1 = LLM(model="TinyLlama/TinyLlama-1.1B-Chat-v1.0")

print("Generating completions with TinyLlama (vLLM)...")
start_time_tinyllama = time.time()
outputs_tinyllama = llm1.generate(prompts, sampling_params)
end_time_tinyllama = time.time()
tinyllama_inference_time = end_time_tinyllama - start_time_tinyllama

# 6. Store and Print results
tinyllama_completions = []
for i, output in enumerate(outputs_tinyllama):
    print(f"\nPrompt {i+1}: {prompts[i]}")
    print(f"Completion (TinyLlama): {output.outputs[0].text}")
    tinyllama_completions.append(output.outputs[0].text)

print(f"\nTinyLlama Inference Time: {tinyllama_inference_time:.2f} seconds")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import time

# Load EleutherAI/pythia-1b using Hugging Face Transformers
print("\nLoading EleutherAI/pythia-1b using Transformers...")
tokenizer = AutoTokenizer.from_pretrained("EleutherAI/pythia-1b")
model = AutoModelForCausalLM.from_pretrained("EleutherAI/pythia-1b")

# Generate completions for each prompt
print("Generating completions with Pythia-1b...")
pythia_completions = []
start_time_pythia = time.time()
for i, prompt in enumerate(prompts):
    inputs = tokenizer(prompt, return_tensors="pt")

    # Generate output based on sampling_params
    # Adapt vLLM's SamplingParams to Hugging Face generate parameters
    outputs = model.generate(
        **inputs,
        max_new_tokens=sampling_params.max_tokens,
        do_sample=True,
        temperature=sampling_params.temperature,
        top_p=sampling_params.top_p
    )

    # Decode and print the completion
    completion = tokenizer.decode(outputs[0], skip_special_tokens=True)
    print(f"\nPrompt {i+1}: {prompt}")
    print(f"Completion (Pythia-1b): {completion}")
    pythia_completions.append(completion)
end_time_pythia = time.time()
pythia_inference_time = end_time_pythia - start_time_pythia

print(f"\nPythia-1b Inference Time: {pythia_inference_time:.2f} seconds")
```

```python
# Load anas-awadalla/mpt-1b-redpajama-200b using Hugging Face Transformers
from transformers import AutoModelForCausalLM, AutoTokenizer
import time

print("\nLoading anas-awadalla/mpt-1b-redpajama-200b using Transformers...")
tokenizer_mpt = AutoTokenizer.from_pretrained("anas-awadalla/mpt-1b-redpajama-200b", trust_remote_code=True)
model_mpt = AutoModelForCausalLM.from_pretrained("anas-awadalla/mpt-1b-redpajama-200b", trust_remote_code=True, dtype="auto")

# Generate completions for each prompt
print("Generating completions with MPT model...")
mpt_completions = []
start_time_mpt = time.time()
for i, prompt in enumerate(prompts):
    inputs_mpt = tokenizer_mpt(prompt, return_tensors="pt")

    # Generate output based on sampling_params
    outputs_mpt = model_mpt.generate(
        **inputs_mpt,
        max_new_tokens=sampling_params.max_tokens,
        do_sample=True,
        temperature=sampling_params.temperature,
        top_p=sampling_params.top_p
    )

    # Decode and print the completion
    completion_mpt = tokenizer_mpt.decode(outputs_mpt[0], skip_special_tokens=True)
    print(f"\nPrompt {i+1}: {prompt}")
    print(f"Completion (MPT): {completion_mpt}")
    mpt_completions.append(completion_mpt)
end_time_mpt = time.time()
mpt_inference_time = end_time_mpt - start_time_mpt

print(f"\nMPT Inference Time: {mpt_inference_time:.2f} seconds")
```

```python
import pandas as pd

# Create a DataFrame for comparison
comparison_df = pd.DataFrame({
    'Prompt': prompts,
    'TinyLlama (vLLM)': tinyllama_completions,
    'Pythia-1b (Transformers)': pythia_completions,
    'MPT-1b (Transformers)': mpt_completions
})

print("\n--- Model Completion Comparison ---")
display(comparison_df)

print("\n--- Model Inference Time Comparison ---")
print(f"TinyLlama Inference Time: {tinyllama_inference_time:.2f} seconds")
print(f"Pythia-1b Inference Time: {pythia_inference_time:.2f} seconds")
print(f"MPT-1b Inference Time: {mpt_inference_time:.2f} seconds")
```
